These are examples of **Exceptions**. When some errors such as the above occur, the python interpreter *throws* an appropriate exception so that the user is made aware and may deal with the exception (because the interpreter per se has no way to proceed at this point)

You can *catch* these exceptions and *handle* them

```python
try:
    x = [2,4]
    print(x[2])
    print("I should never get here")
except IndexError:
    print("This is how you can catch and Index Error")
```

```python
try:
    x = [2,4]
    x = 3/0
    print("I should never get here")
except:
    print("This is how you can catch and Index Error")
```

```python
try:
    x = [2,4]
    x = 3/0
    print("I should never get here")
except IndexError:
    print("This is how you can catch and Index Error")
except Exception:
    print("Something other than a Index Error")
```

You can actually make use of this ability to catch and handle exceptions as follows:

```python
c = {}
s = input()
for w in s.split():
    try:
        c[w] = c[w] + 1
    except KeyError:
        c[w] = 1
print(c)
```

You can catch multiple exceptions in the same try

```python
while True:
    try:
        x = int(input())
        print(3/x)
    except ValueError:
        print("That is not an Integer. Please enter a valid Integer")
    except ZeroDivisionError:
        print("0 entered")
    else:
        break
```

We can make our own Exceptions too as follows

```python
class MyException(Exception):     # Just a trivial Exception class
    pass

try:
    print(" I am just going to throw a MyException ")
    raise MyException(" and catch it")
except MyException as z:
    print(z)
```

A more meaningful application is for instance in **priorityQueue** where we just aborted if asked **deletemax** from an empty state. Unlike the earlier code, where we returned **None**, this forces the user of this class to know that cannot deletemax from an empty queue and must write a handler in case his/her code may make such a call.

```python
class priorityQ():

    def __init__(self): 
        self.L = []
        self.N = 0

    def insert(self,v):
        self.L.append(v)
        self.N = self.N + 1


    def deletemax(self):
        if self.N == 0:       
            raise self.EmptyQException("No values to Delete")
        maxpos = 0
        i = 1
        while (i < self.N):
            if self.L[i] > self.L[maxpos]:
                maxpos  = i
            i = i + 1
        self.N = self.N - 1
        return(self.L.pop(maxpos))
    
    class EmptyQException(Exception):
        pass
```

```python
x = priorityQ()
x.insert(10)
x.insert(2)
print(x.deletemax())
print(x.deletemax())
print(x.deletemax())
```

```python
x = priorityQ()
for i in range(10):
    x.insert(i)
while True:
    try:
        print(x.deletemax())
    except priorityQ.EmptyQException:
        print("Finished. The Q is empty")
        break
```

The **try** construct also permits a **finally** case which is executed no matter whether or which exception is raised. This is usually used to clean up (for instance, close the files that were open etc. We will see this later when we discuss file based input and output)

```python
while True:
    try:
        x = int(input())
        print(3/x)
    except ValueError:
        print("That is not an Integer. Please enter a valid Integer")
    except ZeroDivisionError:
        print("0 entered")
        break
    finally:
        print("This will be printed, always")
```
