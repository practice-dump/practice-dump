```python
from datasets import load_dataset
from datasets import Audio

minds = load_dataset("PolyAI/minds14", name="en-AU", split="train")
minds = minds.cast_column("audio", Audio(sampling_rate=16_000))
```

## Audio Classification Pipeline

```python
from transformers import pipeline

classifier = pipeline(
    "audio-classification",
    model="anton-l/xtreme_s_xlsr_300m_minds14", ## FineTuned on Minds14 dataset
)
```

 This pipeline expects the audio data as a NumPy array. The preprocessing of the raw audio data will be conveniently handled for us by the pipeline.

```python
example = minds[5]
```

```python
classifier(example["audio"]["array"])
```

```python
id2label = minds.features["intent_class"].int2str
id2label(example["intent_class"])
```
